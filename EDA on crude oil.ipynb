#### Q1: average  - what is the average rate of crude oil?

```python
sns.set_style("whitegrid")
avg=df.price.mean()
```

```python
print(" the average crude oil rate is {0:.2f}".format(avg))
```

#### Q2:  max  - In which year the rate of crude oil is high?

```python
Maxi=df.price.max()
high=df[df.price==Maxi].year
high
```

```python
print('The highest price of crude oil is {0}RS on {1}'.format(Maxi,int(high)))
```

#### Q3:  In which year the average crude oil was minimum?

```python
df_year=df.groupby('year')[['price','year']].mean()
df_year.head(10) 
```

```python
low=df_year.min()
```

```python
y=df_year[df_year.price==low].year
```

```python
print("In {} the crude oil average rate was minimum. ".format(int(y)))
```

#### Q4:In which year the average crude oil was maximum?

```python
high=df_year.max()
```

```python
plt.plot(df_year.year,df_year.price,'X-g');
plt.xlabel('Year')
plt.ylabel("price")
plt.legend(['avg price']);
plt.title('The average crude oil ')
```

```python
x=df_year[df_year.price==high].year
```

```python
print("In {} the crude oil average rate was maximum. ".format(int(x)))
```

#### Q5: when was the crude rate was high ...weekend or workdays?
